```python
"""
necessary imports
"""
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import warnings
```

```python
mainfolder = input('enter folder name: ') #project_files\Stream_ASD_001'
dict_sleep = {}

for folder in sorted(os.listdir(mainfolder)):
    #print(folder)
    if folder.endswith('.png') or folder.endswith('.csv') or folder.endswith('.txt') or folder.endswith('.xlsx') or folder.endswith('.pdf')or folder.endswith('.edf'):
            continue
    for subfolder in os.listdir(os.path.join(mainfolder, folder)):
        if subfolder.startswith('Night'):
            print(subfolder)
            for file in os.listdir(os.path.join(mainfolder, folder, subfolder)):
                if file.endswith('_report.csv'):
                    df_sleep = pd.read_csv(os.path.join(mainfolder, folder, subfolder, file))
                    mask = (df_sleep['record'] == 'tst') | (df_sleep['record'] == 'sol') | (df_sleep['record'] == 'waso') | (df_sleep['record'] == 'n1_duration') | (df_sleep['record'] == 'n2_duration') | (df_sleep['record'] == 'n3_duration') | (df_sleep['record'] == 'rem_duration') | (df_sleep['record'] == 'nrem_duration') | (df_sleep['record'] == 'n1_percentage') | (df_sleep['record'] == 'n2_percentage') | (df_sleep['record'] == 'n3_percentage') | (df_sleep['record'] == 'rem_percentage') | (df_sleep['record'] == 'nrem_percentage') | (df_sleep['record'] == 'awakenings') | (df_sleep['record'] == 'sleep_efficiency') 
                    indices = df_sleep.index[mask].tolist()
                    feat_arr = np.array(df_sleep[df_sleep.columns[1]].iloc[indices])
                    dict_sleep[folder] = feat_arr

dict_sleep
```

```python
import matplotlib.pyplot as plt

# Convert string values in dict_sleep to floats
for day in dict_sleep:
    dict_sleep[day] = [float(value) for value in dict_sleep[day]]

# Extract the required data from the dictionary
days = sorted(dict_sleep.keys())  # Sorting days for consistent plotting
tst_values = [dict_sleep[day][0] for day in days]  # Total Sleep Time (TST) - first value
sol_values = [dict_sleep[day][1] for day in days]  # Sleep Onset Latency (SOL) - second value
waso_values = [dict_sleep[day][2] for day in days]  # Wake After Sleep Onset (WASO) - third value

# Create a figure
plt.figure(figsize=(10, 8))

# Plot TST
plt.plot(days, tst_values, label='Total Sleep Time (TST)', marker='o', linestyle='-', color='blue')

# Plot SOL
plt.plot(days, sol_values, label='Sleep Onset Latency (SOL)', marker='o', linestyle='-', color='orange')

# Plot WASO
plt.plot(days, waso_values, label='Wake After Sleep Onset (WASO)', marker='o', linestyle='-', color='green')

# Set y-axis limits and ticks
plt.ylim(0, 500)  # Set y-axis limits between 0 and 500
plt.yticks(range(0, 501, 100))  # Set y-ticks with a step of 100

# Set plot labels and title
plt.title('Sleep Metrics Across Days', fontsize=16)
plt.xlabel('Days', fontsize=14)
plt.ylabel('Time (Minutes)', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Show legend
plt.legend(fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
